# Linear model on bmi and age

```python
y=df.charges
```

```python
x=df.iloc[:,[0,2]]
```

```python
x
```

```python
model=LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x,y)
```

```python
model.score(x_test,y_test)
```
